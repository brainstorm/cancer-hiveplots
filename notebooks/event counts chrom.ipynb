```python
import pandas as pd
import numpy as np

import feather
```

```python
dp10 = pd.read_csv('data/ohoffman_brain/sv_plots/brain_tumor-svs-full-dp10.csv')
dp25 = pd.read_csv('data/ohoffman_brain/sv_plots/brain_tumor-svs-full-dp25.csv')

apgi = pd.read_csv('data/bndgenotype/sv_plots/APGI-svs-full-dp10.csv')

empty_chrom_df = pd.DataFrame([x for x in range(23)])
```

```python
#dp10.head()
apgi.head()
```

```python
intra_dp10 = dp10[dp10["chrom1"] == dp10["chrom2"]]
inter_dp10 = dp10[dp10["chrom1"] != dp10["chrom2"]]

intra_apgi = apgi[apgi['chrom1'] == apgi['chrom2']]
inter_apgi = apgi[apgi['chrom1'] != apgi['chrom2']]
```

```python
grouped_intra = intra_dp10.groupby("chrom1").size()
grouped_inter = inter_dp10.groupby("chrom1").size()
```

```python
apgi1953_intra_unstable_lumpy = intra_apgi[(intra_apgi["sample"] == "APGI1953_Tumor") & (intra_apgi["caller"] == "lumpy")]
apgi1955_intra_focal_lumpy = intra_apgi[(intra_apgi["sample"] == "APGI1955_Tumor") & (intra_apgi["caller"] == "lumpy")]
apgi2049_intra_stable_lumpy = intra_apgi[(intra_apgi["sample"] == "APGI2049_Tumor") & (intra_apgi["caller"] == "lumpy")]

apgi1953_inter_unstable_lumpy = inter_apgi[(inter_apgi["sample"] == "APGI1953_Tumor") & (inter_apgi["caller"] == "lumpy")]
apgi1955_inter_focal_lumpy = inter_apgi[(inter_apgi["sample"] == "APGI1955_Tumor") & (inter_apgi["caller"] == "lumpy")]
apgi2049_inter_stable_lumpy = inter_apgi[(inter_apgi["sample"] == "APGI2049_Tumor") & (inter_apgi["caller"] == "lumpy")]
```

```python
size_apgi1953 = apgi1953_intra_unstable_lumpy.groupby("chrom1").size()
size_apgi1955 = apgi1955_intra_focal_lumpy.groupby("chrom1").size()
size_apgi2049 = apgi2049_intra_stable_lumpy.groupby("chrom1").size()

size_apgi1953_inter = apgi1953_inter_unstable_lumpy.groupby("chrom1").size()
size_apgi1955_inter = apgi1955_inter_focal_lumpy.groupby("chrom1").size()
size_apgi2049_inter = apgi2049_inter_stable_lumpy.groupby("chrom1").size()
```

```python
apgi_intra_lumpy = pd.DataFrame({"unstable": size_apgi1953, "focal": size_apgi1955, "stable": size_apgi2049}, columns = ["unstable", "focal", "stable"])
apgi_inter_lumpy = pd.DataFrame({"unstable": size_apgi1953_inter, "focal": size_apgi1955_inter, "stable": size_apgi2049_inter}, columns = ["unstable", "focal", "stable"])
```

```python
apgi_inter_lumpy.fillna(0, inplace=True)
```

```python
apgi_intra_lumpy.to_excel("APGI_intra_counts_chrom.xlsx")
apgi_inter_lumpy.to_excel("APGI_inter_counts_chrom.xlsx")
```
